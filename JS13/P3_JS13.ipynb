## Praktikum 3:

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf

# Contoh dataset (buat dummy data)
data = pd.DataFrame({
    'luas': [50, 60, 70, 80, 90],
    'harga': [500, 600, 700, 800, 900]
})

X = data[['luas']]
y = data[['harga']]

# Normalisasi
scaler = StandardScaler()
X = scaler.fit_transform(X)
y = scaler.fit_transform(y)

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation='relu', input_shape=(1,)),
    tf.keras.layers.Dense(1)
])

model.compile(optimizer='adam', loss='mse')
model.fit(X_train, y_train, epochs=100)

# Evaluasi
print("Prediksi:", model.predict(X_test))
```

### Tugas 2

- Ubah learning rate.
- Bandingkan hasil loss.

**Learning Rate 0.1**

```python
model_lr1 = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation='relu', input_shape=(1,)),
    tf.keras.layers.Dense(1)
])

model_lr1.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.1),
                  loss='mse')

history1 = model_lr1.fit(X_train, y_train, epochs=100, verbose=0)

loss1 = history1.history['loss'][-1]
print("Loss LR=0.1:", loss1)
```

**Learning rate 0.01**

```python
model_lr2 = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation='relu', input_shape=(1,)),
    tf.keras.layers.Dense(1)
])

model_lr2.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
                  loss='mse')

history2 = model_lr2.fit(X_train, y_train, epochs=100, verbose=0)

loss2 = history2.history['loss'][-1]
print("Loss LR=0.01:", loss2)
```

**Learning Rate 0.001**

```python
model_lr3 = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation='relu', input_shape=(1,)),
    tf.keras.layers.Dense(1)
])

model_lr3.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
                  loss='mse')

history3 = model_lr3.fit(X_train, y_train, epochs=100, verbose=0)

loss3 = history3.history['loss'][-1]
print("Loss LR=0.001:", loss3)
```

**Learning Rate 0.0001**

```python
model_lr4 = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation='relu', input_shape=(1,)),
    tf.keras.layers.Dense(1)
])

model_lr4.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
                  loss='mse')

history4 = model_lr4.fit(X_train, y_train, epochs=100, verbose=0)

loss4 = history4.history['loss'][-1]
print("Loss LR=0.0001:", loss4)
```

Berdasarkan hasil tersbeut, dapat disimpulkan bahwa learning rate paling tepat untuk kasus ini adalah 0.01 karena menghasilkan loss yang lebih kecil dari yang lainnya yaitu sebeesar 0.0044..

## Praktikum berikut akan menggunakan data Boston untuk memprediksi harga rumah.

```python
# MLP regresi (Keras)
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing   
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

# 1. Load
data = fetch_california_housing()
X = data.data; y = data.target

# 2. Preprocess
scaler = StandardScaler(); Xs = scaler.fit_transform(X)
X_train, X_val, y_train, y_val = train_test_split(Xs, y, test_size=0.2, random_state=42)

# 3. Build model
model = Sequential([
Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
Dense(32, activation='relu'),
Dense(1)
])
model.compile(optimizer=Adam(1e-3), loss='mse', metrics=['mae'])

# 4. Train
h = model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=200, batch_size=32, verbose=0)

# 5. Plot
plt.figure(figsize=(10,4))
plt.subplot(1,2,1); plt.plot(h.history['loss'], label='train_loss'); plt.plot(h.history['val_loss'], label='val_loss'); plt.legend(); plt.title('MSE')
plt.subplot(1,2,2); plt.plot(h.history['mae'], label='train_mae'); plt.plot(h.history['val_mae'], label='val_mae'); plt.legend(); plt.title('MAE')
plt.show()

from sklearn.metrics import mean_squared_error
pred = model.predict(X_val)
print('RMSE:', np.sqrt(mean_squared_error(y_val, pred)))
```
